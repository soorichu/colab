```
def method():
  print("start method()")
  val = []
  for x in range(0, 5):
    val.append(x**2)
  return val

out = method()
```

method()의 경우 실행하면 "start method()"가 바로 실행된다.

```
for i in range(0, 5):
  print(out[i])
```

# 제너레이터
제너레이터는 딥러닝에서 학습용 데이터의 Batch를 만들 때 많이 사용된다. 위의 return을 사용한 함수 method()는 성능 문제가 있다. 만약 range(0, 100000) 같이 범위가 매우 커진다면 메모리 공간도 부족하고 성능이 저하된다. 
제너레이터를 사용하면 메모리 문제와 성능 저하 문제를 해결해준다. yield 키워드를 사용하여 제너레이터를 이용할 수 있다. 

```
def generator(n):
  print("Start generator()")
  for x in range(0, n):
    yield x**2
    
gen = generator(100)
```

geherator(100)실행 시 보다 시피 "Start generator()"가 실행되지 않는다. 계산 부분이 실행되지 않은 채로 제너레이터가 선언만 되어 있는 상태에다. 
다음과 같이 next()를 사용해야 실제 계산이 실행된다.

```
for i in range(100):
  print(next(gen))
```

next를 호출할 때마다 yield 까지만 실행된다. 그러므로 메모리 공간도 아낄 수 있고 성능 저하도 막을 수 있다.

# zip이 들어간 for loop
보통 zip는 압축 파일의 확장자에서 많이 볼 수 있다. 파이썬에도 zip() 함수가 있는데 압축 기능은 아니다. 압축은 여러 파일을 한 가지 파일로 묶어준다고 생각할 수 있다. 파이썬의 zip()는 동일한 개수로 이루어진 데이터타입을 묶어주는 역할을 한다.

```
items1 = [[1,2], [3, 4], [5, 6]]
items2 = [['A', 'B'], ['C', 'D'], ['E','F']]
print(items1)
print(items2)
```

```
for dig, char in zip(items1, items2):
  print(dig, char)
```

# 다변수 함수 그래프

```
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
```

```
fig = plt.figure(figsize=(5, 5))
plt.contourf(X, Y, Z, cmap='coolwarm')
plt.grid()
plt.axis('equal')
plt.show()
```
