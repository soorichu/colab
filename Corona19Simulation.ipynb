```
E = 5.2               #코로나19 평균 잠복기(고정)
I = 6325              #2020.3.20. 00시 질병관리본부 치료 중
R = 2233              #2020.3.20. 00시 질병관리본부 완치
D = 94                #2020.3.20. 00시 질병관리본부 사망
S = 51470000          #2017년 대한민국 인구수
alpha = 327000/S/365  #통계청 발표 2018년 출생아수/인구수/365
beta = 298900/S/365   #통계청 발표 2018년 사망자수/인구수/365
dD = 3
delta = dD/I
dR = 286
c = (dR+beta*R)/I
dS = alpha-beta-87               #2020.3.20. 00시 질병관리본부 전일대비 증가한 확진자수를 뺌
a = (alpha*S-beta*S-dS)/(S*I)
dI = 87                #2020.3.20. 00시 질병관리본부 전일대비 증가한 확진자수
b = (beta*I+c*I+delta*I + dI)/E
MAX_TIME = 365*5       #시간 t는 1일
#print(a, alpha, beta, delta)
for t in range(MAX_TIME):
  print(t, "DAYS : ", I/S, I, R, D)
  tempS = S + (alpha*S - beta*S - a*S*I)
  #tempE = E + (alpha*S*I - beta*E - b*E)
  tempI = I + (b*E - beta*I - c*I - delta*I)
  tempR = R + (c*I - beta*R)
  tempD = D + (delta*I)
  dS = int(tempS) - S; dI = int(tempI) - I; dR = int(tempR) - R;
  S = int(tempS); I = int(tempI); R = int(tempR); D = int(tempD) 
  #E = tempE
```
